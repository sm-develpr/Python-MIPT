# 1. Абстракции

Напиши абстрактный класс `Organism`. От него будут наследоваться другие классы.

В нём должен быть:

Concrete Methods:

- `property` - `alive`, который проверяет, что текущий возраст не больше, чем максимальный и энергия не отрицательная. При попытке изменения (сеттер), нужно сделать возраст больше, чем максимальный, на 1.
- Мэджик метод, позволяющий использовать экземпляры в if-конструкциях (мэджик метод `__bool__`, связанный с `alive`).
- Инициализатор (`__init__`), который проставляет атрибуты экземпляра: размер популяции `population_size = 0`, текущий возраст `current_age = 0`, `current_energy = 0` и `id` в зависимости от текущей глобальной переменной `ID_COUNTER`. Последнюю надо увеличить на 1, чтобы последующие ID отличались.
- Метод класса (`classmethod`) `spawn`, принимающий на вход `population_size`, `current_age` и фиксирующий их для объекта.
- Метод `reproduce`, который возвращает новый объект данного класса, в котором `current_age = 0`, `current_energy` - меньше в `REPRODUCE_NEW_COEF` раз, округлённых вниз; population_size - меньше в `REPRODUCE_POPULATION_COEF` раз, округлённых вниз. Кроме того, энергия изначального объекта понижается в `REPRODUCE_OLD_COEF` раз и логгирует в поток вывода сообщение с ID тех, кто размножился и тех, кто появился.


Abstract Methods:

- Мэджик метод, отвечающий за репрезентацию объекта (str/repr)
- Метод `replenish_energy`, отвечающий за пополнение энергии



```python
from abc import ABC, abstractmethod
from math import floor
from typing import override

ID_COUNTER = 1

class Organism(ABC):
    @property
    def alive(self):
        return self.current_age <= self.max_age and self.current_energy >= 0

    @alive.setter
    def alive(self, value):
        if not value:
            self.current_energy = self.max_age + 1
        else:
            self.current_energy = value

    def __bool__(self):
        return self.alive

    def __init__(self):
        global ID_COUNTER
        self.population_size = 0
        self.current_age = 0
        self.current_energy = 0
        self.id = ID_COUNTER
        ID_COUNTER += 1
        self.REPRODUCE_POPULATION_COEF = 10
        self.REPRODUCE_NEW_COEF = 10
        self.REPRODUCE_POPULATION_COEF = 10

    @classmethod
    def spawn(cls, population_size, current_age):
        organism = cls()
        organism.population_size = population_size
        organism.current_age = current_age
        return organism

    def reproduce(self):
        obj = self.__class__.spawn(floor(self.population_size / self.REPRODUCE_POPULATION_COEF), 0)
        obj.current_energy = floor(self.current_energy / self.REPRODUCE_NEW_COEF)
        self.current_energy /= self.REPRODUCE_NEW_COEF

        print(f"log: [ID] {self.id} was reproduced {obj.id}")
        return obj

    @abstractmethod
    def __str__(self):
        pass
    @abstractmethod
    def __repr__(self):
        pass
    @abstractmethod
    def replanish_energy(self, val):
        pass

# Что то я не понимаю зачем тут какие то не понятные контсанты которые должны то ли уже быть объявлены в глобале, то ли внутри класса(???)
```

- `replenish_energy(eaten_plant_energy)`, который увеличивает количество энергии на `eaten_plant_energy`.


Для Plant:

- Инициализатор, который проставляет `max_age` равным `MAX_PLANT_AGE`

```python
MAX_HERBIVIOR_AGE = 20
MAX_PLANT_AGE = 300

class Herbivore(Organism):
    def __init__(self):
        super().__init__()
        self.max_age = MAX_HERBIVIOR_AGE

    @override
    def replanish_energy(self, eating_plant_energy):
        self.current_energy += eating_plant_energy

    @override
    def __str__(self):
        pass
    @override
    def __repr__(self):
        pass

class Plant(Organism):
    def __init__(self):
        super().__init__()
        self.max_age = MAX_PLANT_AGE

    @override
    def replanish_energy(self, PLANT_ENERGY_COEF):
        self.current_energy += PLANT_ENERGY_COEF

    @override
    def __str__(self):
        pass
    @override
    def __repr__(self):
        pass

```
