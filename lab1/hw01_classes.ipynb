# 1. Абстракции

Напиши абстрактный класс `Organism`. От него будут наследоваться другие классы.

В нём должен быть:

Concrete Methods:

- `property` - `alive`, который проверяет, что текущий возраст не больше, чем максимальный и энергия не отрицательная. При попытке изменения (сеттер), нужно сделать возраст больше, чем максимальный, на 1.
- Мэджик метод, позволяющий использовать экземпляры в if-конструкциях (мэджик метод `__bool__`, связанный с `alive`).
- Инициализатор (`__init__`), который проставляет атрибуты экземпляра: размер популяции `population_size = 0`, текущий возраст `current_age = 0`, `current_energy = 0` и `id` в зависимости от текущей глобальной переменной `ID_COUNTER`. Последнюю надо увеличить на 1, чтобы последующие ID отличались.
- Метод класса (`classmethod`) `spawn`, принимающий на вход `population_size`, `current_age` и фиксирующий их для объекта.
- Метод `reproduce`, который возвращает новый объект данного класса, в котором `current_age = 0`, `current_energy` - меньше в `REPRODUCE_NEW_COEF` раз, округлённых вниз; population_size - меньше в `REPRODUCE_POPULATION_COEF` раз, округлённых вниз. Кроме того, энергия изначального объекта понижается в `REPRODUCE_OLD_COEF` раз и логгирует в поток вывода сообщение с ID тех, кто размножился и тех, кто появился.


Abstract Methods:

- Мэджик метод, отвечающий за репрезентацию объекта (str/repr)
- Метод `replenish_energy`, отвечающий за пополнение энергии



```python
from abc import ABC, abstractmethod
from math import floor
from typing import override

ID_COUNTER = 1

class Organism(ABC):
    @property
    def alive(self):
        return self.current_age <= self.max_age and self.current_energy >= 0

    @alive.setter
    def alive(self, value):
        if not value:
            self.current_age = self.max_age + 1
    def __bool__(self):
        return self.alive

    def __init__(self):
        print("init organism")
        global ID_COUNTER
        self.population_size = 0
        self.current_age = 0
        self.current_energy = 0
        self.id = ID_COUNTER
        ID_COUNTER += 1
        self.REPRODUCE_POPULATION_COEF = 10
        self.REPRODUCE_NEW_COEF = 10
        self.REPRODUCE_POPULATION_COEF = 10

    @classmethod
    def spawn(cls, population_size, current_age):
        print("Start spawn")
        organism = cls()
        organism.population_size = population_size
        organism.current_age = current_age
        return organism

    def reproduce(self):
        obj = self.__class__.spawn(floor(self.population_size / self.REPRODUCE_POPULATION_COEF), 0)
        obj.current_energy = floor(self.current_energy / self.REPRODUCE_NEW_COEF)
        self.current_energy //= self.REPRODUCE_NEW_COEF

        print(f"log: [ID] {self.id} was reproduced {obj.id}")
        return obj

    @abstractmethod
    def __str__(self):
        pass
    @abstractmethod
    def __repr__(self):
        pass
    @abstractmethod
    def replanish_energy(self, val):
        pass

```

Напиши следующие Mixin'ы:

- `ShortLifespan` - уменьшает максимальный возраст организма в `SHORT_LIFESPAN_COEF` раз, но увеличивает прибавку энергии в `SHORT_LIFESPAN_ENERGY_COEF` раз

- `LongLifespan` - обратное предыдущему. Добавь соответствующие константы.

- `HighFertilityMixin` - увеличивает количество детей в методе `reproduce` в `HIGH_FERTILITY_COEF` раз, но повышает затраты энергии в `reproduce` в `HIGH_FERTILITY_ENERGY_COEF` раз.

- `Toxic` (только для растений) - при поедании такого растения количество животных из данного набора уменьшается в `TOXIC_COEF` раз.

- `IdentifiesToxicPlants` (только для травоядных) - при выборе растения в фазе поедания, если растение токсично, выбирают не есть его.

- `ResilientToToxicPlants` (только для травоядных) - при выборе токсичного растения смертность уменьшена в `RESILIENT_COEF` раз.

```python
SHORT_LIFESPWAN_COEF = 2
SHORT_LIFESPAWN_ENERGY_COEF = 2

class ShortLifespawnMixin:
  def __init__(self, *args, **kwargs):
    super().__init__(*args, **kwargs)
    self.max_age //= SHORT_LIFESPWAN_COEF
    self.current_energy //= SHORT_LIFESPAWN_ENERGY_COEF
  
  def replanish_energy(self, val):
    super().replanish_energy(val * SHORT_LIFESPAWN_ENERGY_COEF)


LONG_LIFESPAWN_COEF = 2
LONG_LIFESPAWN_ENERGY_COEF = 2


class LongLifespawnMixin:
  def __init__(self, *args, **kwargs):
    super().__init__(*args, **kwargs)
    self.max_age *= LONG_LIFESPAWN_COEF
    self.current_energy *= LONG_LIFESPAWN_ENERGY_COEF
  
  def replanish_energy(self, val):
    super().replanish_energy(val // LONG_LIFESPAWN_ENERGY_COEF)



HIGH_FERTILITY_COEF = 2
HIGH_FERTILITY_ENERGY_COEF = 2

class HighFertilityMixin:
  def reproduce(self):
    self.REPRODUCE_NEW_COEF *= self.HIGH_FERTILITY_ENERGY_COEF
    for _ in range(self.HIGH_FERTILITY_COEF):
      if not self.alive or self.current_energy <= 0 or self.population_size <= 0:
        break
      super().reproduce()
    self.REPRODUCE_NEW_COEF //= self.HIGH_FERTILITY_ENERGY_COEF
  

class ToxicMixin:
  TOXIC_COEF = 2
  @property
  def is_toxic(self) -> bool:
    return True

class IdentifiesToxicPlantsMixin:
  def eat(self, plant):
    if isinstance(plant, ToxicMixin):
      return
    return super().eat(plant)


class ResilientToToxicPlants:
  RESILIENT_COEF = 2
  def eat(self, plant):
    if isinstance(plant, ToxicMixin):
      self.population_size = max(1, floor(self.population_size / max(1, plant.TOXIC_COEF // self.RESILIENT_COEF)))
      self.replanish_energy(plant.current_energy)
      return
    return super().eat(plant)
  



```

Перекопируй код для абстрактного класса из п. 1 данной работы и добавь к нему метод `mutate`: с вероятностью `MUTATION_POSSIBILITY` процентов он выбирает какой-то из mixin'ов, который у него ещё не стоит в родителях, и добавляет его в родителей; иначе - он выбирает одного из родителей и удаляет его из кортежа родителей (`__bases__`).


Подумай, как это реализовать так, чтобы сам метод был общий и для `Plant`, и для `Herbivore`, но при этом выбор возможных Mixin'ов был разный. При надобности их тоже скопируй ниже и поправь в них код.


Также важно! в методе `mutate` должно быть логгирование (`print` с описанием произошедшего), чтобы было понятно, кто мутировал.

```python
from abc import ABC, abstractmethod
from math import floor
from random import randint
from typing import override

class Organism(ABC):
  MUTATION_POSSIBILITY = 30
  AVAILABLE_MIXINS = ()

  @property
  def alive(self):
    return self.current_age <= self.max_age and self.current_energy >= 0

  @alive.setter
  def alive(self, value):
    if not value:
      self.current_age = self.max_age + 1

  def __bool__(self):
    return self.alive

  def __init__(self):
    print("init organism")
    global ID_COUNTER
    self.population_size = 0
    self.current_age = 0
    self.current_energy = 0
    self.id = ID_COUNTER
    ID_COUNTER += 1
    self.REPRODUCE_POPULATION_COEF = 10
    self.REPRODUCE_NEW_COEF = 10
    self.REPRODUCE_POPULATION_COEF = 10

  @classmethod
  def spawn(cls, population_size, current_age):
    print("Start spawn")
    organism = cls()
    organism.population_size = population_size
    organism.current_age = current_age
    return organism

  def reproduce(self):
    obj = self.__class__.spawn(floor(self.population_size / self.REPRODUCE_POPULATION_COEF), 0)
    obj.current_energy = floor(self.current_energy / self.REPRODUCE_NEW_COEF)
    self.current_energy //= self.REPRODUCE_NEW_COEF

    print(f"log: [ID] {self.id} was reproduced {obj.id}")
    return obj

  def mutate(self):
    cls = self.__class__

    available = tuple(getattr(cls, "AVAILABLE_MIXINS", ()))
    current_mixins = [b for b in cls.__bases__ if b in available]
    missing_mixins = [m for m in available if m not in cls.__bases__]

    roll = randint(1, 100)
    if roll <= self.MUTATION_POSSIBILITY:
      if not missing_mixins:
        print(f"log: [ID] {self.id} mutate: no mixins to add for {cls.__name__}")
        return

      mixin = missing_mixins[randint(0, len(missing_mixins) - 1)]
      NewCls = type(
        f"{cls.__name__}_MutAdd_{mixin.__name__}_{self.id}",
        (mixin, cls),
        {}
      )
      new_obj = NewCls.__new__(NewCls)
      new_obj.__dict__ = self.__dict__.copy()
      print(f"log: [ID] {self.id} mutate: {cls.__name__} added mixin {mixin.__name__})")
      return new_obj

    if not current_mixins:
      print(f"log: [ID] {self.id} mutate: no mixins to remove for {cls.__name__} (roll={roll})")
      return

    remove_mixin = current_mixins[randint(0, len(current_mixins) - 1)]
    kept = [m for m in current_mixins if m is not remove_mixin]
    NewCls = type(
      f"{cls.__name__}_MutRem_{remove_mixin.__name__}_{self.id}",
      tuple(kept + [cls]),
      {}
    )
    new_obj = NewCls.__new__(NewCls)
    new_obj.__dict__ = self.__dict__.copy()
    print(f"log: [ID] {self.id} mutate: {cls.__name__} removed mixin {remove_mixin.__name__}")
    return new_obj

  @abstractmethod
  def __str__(self):
    pass

  @abstractmethod
  def __repr__(self):
    pass

  @abstractmethod
  def replanish_energy(self, val):
    pass

MAX_HERBIVIOR_AGE = 20
MAX_PLANT_AGE = 300

class Herbivore(Organism):
    AVAILABLE_MIXINS = (
        ShortLifespawnMixin,
        LongLifespawnMixin,
        HighFertilityMixin,
        IdentifiesToxicPlantsMixin,
        ResilientToToxicPlants,
    )
    def __init__(self):
        print("init herbivore")
        super().__init__()
        self.max_age = MAX_HERBIVIOR_AGE

    @override
    def replanish_energy(self, eating_plant_energy):
        self.current_energy += eating_plant_energy

    def eat(self, plant: "Plant"):
        self.replanish_energy(plant.current_energy)
        if isinstance(plant, ToxicMixin):
            self.population_size = max(1, floor(self.population_size / TOXIC_COEF))

    @override
    def __str__(self):
        return f"Herbivore(age={self.age}, energy={self.current_energy})"
    @override
    def __repr__(self):
        return self.__str__()


class Plant(Organism):
  AVAILABLE_MIXINS = (
    ShortLifespawnMixin,
    HighFertilityMixin,
    LongLifespawnMixin,
    ToxicMixin,
  )

  def __init__(self):
    print("init plant")
    super().__init__()
    self.max_age = MAX_PLANT_AGE

  @override
  def replanish_energy(self, PLANT_ENERGY_COEF):
    self.current_energy += PLANT_ENERGY_COEF

  @override
  def __str__(self):
    return f"Plant(age={self.current_age}, energy={self.current_energy})"

  @override
  def __repr__(self):
    return self.__str__()
```

Каждый из этих трёх действий должно корректно логироваться (чтобы мы понимали, что происходит в симуляции).


- метод `play_game(n)` - который запускает игру и делает последовательно `n` циклов жизни (запусков `correct_age_energy` и `cycle_through_actions`).

```python
class Game:

  HERBIVORE_ENERGY_CONSUMPTION = 5
  PLANT_ENERGY_CONSUMPTION = 1
  PLANT_REPLENISH_ENERGY = 3
  PLANT_EATEN_COEF = 2
  PLANT_ENERGY_EATEN_COEF = 2

  def __init__(self):
    self.animals = [
      Herbivore.spawn(100, 0),
      Herbivore.spawn(50, 0),
    ]
    self.plants = [
      Plant.spawn(100, 0),
      Plant.spawn(50, 0),
    ] 

    for _ in self.animals:
      _.current_energy = 100
    for _ in self.plants:
      _.current_energy = 200
  
  def __alive_animals(self):
    return [a for a in self.animals if a.alive]
  def __alive_plants(self):
    return [p for p in self.plants if p.alive]
  
  def correct_age_energy(self):
    for a in self.__alive_animals():
      a.current_energy -= self.HERBIVORE_ENERGY_CONSUMPTION
      a.current_age += 1

      if not a.alive:
        a.alive = False
        print(f"log: [ID] {a.id} died (Herbivore) age={a.current_age}/{a.max_age}, energy={a.current_energy}")

    for p in self.__alive_plants():
      p.current_energy -= self.PLANT_ENERGY_CONSUMPTION
      p.current_age += 1

      if not p.alive:
        p.alive = False
        print(f"log: [ID] {p.id} died (Plant) age={p.current_age}/{p.max_age}, energy={p.current_energy}")
  
  def play_game(self, n):
    for step in range(1, n + 1):
      print(f"Step {step}/{n}")
      self.correct_age_energy()
      self.cycle_through_actions()

      print(f"log: end cycle {step} -> alive animals={len(self.__alive_animals())}, alive plants={len(self.__alive_plants())}")
   
  def cycle_through_actions(self):
    all = self.__alive_animals() + self.__alive_plants()
    for org in all:
      if not org.alive:
        continue

      action = randint(0, 2)  # 0=replenish, 1=reproduce, 2=mutate

      if action == 0:
        if isinstance(org, Herbivore):
          plants = self.__alive_plants()
          if not plants:
            print(f"log: [ID] {org.id} action=replenish - no plants to eat")
            continue

          plant = plants[randint(0, len(plants) - 1)]

          old_pop = plant.population_size
          old_energy = plant.current_energy

          eaten_pop = max(1, floor(plant.population_size / self.PLANT_EATEN_COEF))
          eaten_energy = max(1, floor(plant.current_energy / self.PLANT_ENERGY_EATEN_COEF))

          plant.population_size = max(0, plant.population_size - eaten_pop)
          plant.current_energy = max(0, plant.current_energy - eaten_energy)

          org.replanish_energy(eaten_energy)

          print(
            f"log: [ID] {org.id} action=replenish ate Plant[ID {plant.id}] "
            f"+{eaten_energy} energy; "
            f"plant_pop {old_pop}->{plant.population_size}, plant_energy {old_energy}->{plant.current_energy}"
          )
          continue

        if isinstance(org, Plant):
          before = org.current_energy
          org.replanish_energy(self.PLANT_REPLENISH_ENERGY)
          print(
            f"log: [ID] {org.id} action=replenish grow "
            f"energy {before}->{org.current_energy}"
          )
          continue

      if action == 1:
        # Для HighFertilityMixin reproduce может вернуть None и сам залогировать детей.
        # Для базового Organism.reproduce возвращает объект.
        print(f"log: [ID] {org.id} action=reproduce")
        child = org.reproduce()

        if child is None:
          continue

        if isinstance(child, Herbivore):
          self.animals.append(child)
        elif isinstance(child, Plant):
          self.plants.append(child)

        print(f"log: [ID] {org.id} reproduce -> new {child.__class__.__name__} [ID {child.id}] pop={child.population_size}, energy={child.current_energy}")
        continue

      if action == 2:
        print(f"log: [ID] {org.id} action=mutate")
        mutated = org.mutate()

        if mutated is not org:
          if isinstance(org, Herbivore):
            for i in range(len(self.animals)):
              if self.animals[i] is org:
                self.animals[i] = mutated
                break
          elif isinstance(org, Plant):
            for i in range(len(self.plants)):
              if self.plants[i] is org:
                self.plants[i] = mutated
                break

```
